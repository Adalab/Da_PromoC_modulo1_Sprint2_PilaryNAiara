---------------------------------- PRIMERA FUNCIÓN -----------------------

Esta función debe incluir el código que:

1. Nos muestre en que carpeta estamos trabajando.

2. Cree una carpeta que se llame "aprendiendo-ficheros". ⚠️ Tened en cuenta que si la carpeta ya existe no la podemos crear, nos devolverá un error. Incluye en la función un programa que evite que nos de un error si la carpeta ya existe.

3. Cree otra carpeta que se llame "datos" dentro de la carpeta "aprendiendo-ficheros". En esta carpeta "datos" guardaremos el fichero "saludo.txt" que os habéis descargado.
4. Cambiad el directorio de trabajo a la carpeta "datos". Antes de seguir chequead que estáis trabajando en la carpeta "datos".
5. Cambiad el nombre de la carpeta creada en el punto 2 a "primera-toma-contacto"

💡 Pistas para resolver este ejercicio 💡 Esta función debe incluir el código que:

1. La función tendrá que recibir 3 parámetros:

- El nombre del nombre de la primera carpeta
- El nombre de la carpeta segunda carpeta
- El nombre con el que queramos cambiar el nombre de la primera carpeta creada

2. Para saber si las carpetas ya existen tendréis que usar lstdir(recordad que nos devuelve una lista de ficheros y carpetas).
3. Para poder controlar los errores tendremos que usar un if loop, que si el fichero existe, nos devuelva un mensaje de que el fichero ya existe. En caso de que no exista, los deberéis crear y que la función nos muestre un mensaje de que se ha creado.
4. Tendréis que ir cambiando de directorio para poder crear las carpetas y cambiar sus nombres.


```python
# importar librería para la gestión de carpetas
import os
```

```python
def crear_carpeta(primera_carpeta, segunda_carpeta, cambio_primera_carpeta):
    # para conocer mi ruta absoluta
    mi_ruta = os.getcwd()
    print(f"Nuestro directorio de trabajo es: '{mi_ruta}'")
    
    ruta_primera_carpeta = mi_ruta + '/' + primera_carpeta

    # para conocer el listado de carpetas que hay en la ruta absoluta
    if primera_carpeta in os.listdir(mi_ruta):
        # la primera_carpeta está creada
        print(f"La carpeta '{primera_carpeta}' ya está creada en '{mi_ruta}'")
    else:
        # nos creamos la primera_carpeta
        os.mkdir(ruta_primera_carpeta)
        if primera_carpeta in os.listdir(mi_ruta):
            print(f"La carpeta '{primera_carpeta}' ha sido creada satisfactoriamente en '{mi_ruta}'")
        else:
            print(f"Se ha producido un error al crear la carpeta '{primera_carpeta}'")

    # para crear la segunda_carpeta entramos dentro de la primera_carpeta

    ruta_segunda_carpeta = ruta_primera_carpeta + '/' + segunda_carpeta

    if segunda_carpeta in os.listdir(ruta_primera_carpeta):
        print(f"La carpeta '{segunda_carpeta}' ya está creada en '{ruta_primera_carpeta}'")
    else:
        os.mkdir(ruta_segunda_carpeta)
        if segunda_carpeta in os.listdir(ruta_primera_carpeta):
            print(f"La carpeta '{segunda_carpeta}' ha sido creada satisfactoriamente en '{ruta_primera_carpeta}'")
        else:
            print(f"Se ha producido un error al crear la carpeta '{segunda_carpeta}'")

    # nos descargamos el fichero "saludo.txt" directamente en la carpeta datos
    # solo printeamos que existe el fichero
    saludo = ruta_segunda_carpeta + '/' + 'saludo.txt'
    print(saludo)
    if saludo in os.listdir(ruta_segunda_carpeta):
        print(f"El fichero 'saludo.txt' está creado en: {ruta_segunda_carpeta}")
    else:
        print(f"El fichero 'saludo.txt' no está creado en: {ruta_segunda_carpeta}")

    # nos cambiamos a la segunda carpeta
    os.chdir(ruta_segunda_carpeta)
    
    # para comprobar que estamos en la carpeta segunda_carpeta
    print(f"El directorio en el que estamos es: ", os.getcwd())

    # para cambiar el nombre tenemos que cambiar la ruta a mi_ruta 
    os.chdir(mi_ruta)

    print("Nos cambiamos al directorio: ", os.getcwd())
    print(f"para poder cambiar el nombre de la carpeta '{primera_carpeta}'")

    ruta_cambio_primera_carpeta = mi_ruta + '/' + cambio_primera_carpeta
    # cambiamos el nombre de primera_carpeta por cambio_primera_carpeta
    os.rename(ruta_primera_carpeta, ruta_cambio_primera_carpeta)

    # solo printeamos que hemos renombrado la carpeta
    if cambio_primera_carpeta in mi_ruta:
        print(f"Hemos renombrado '{primera_carpeta}' por '{cambio_primera_carpeta}'")
    else:
        print(f"No hemos renombrado '{primera_carpeta}' por '{cambio_primera_carpeta}'")
```

3. Usar if... else para evitar que se nos pare el código.
4. Para cambiar el fichero podréis usar el comando input para preguntar el usuario donde está el fichero y que se pueda usar la ruta relativa o absoluta.

```python
def leer_fichero(nombre_fichero):
    # para conocer mi ruta absoluta
    mi_ruta = os.getcwd()
    print(f"Nuestro directorio de trabajo es: '{mi_ruta}'")
    
    # en carpeta nos guardamos la lista de mi_ruta
    carpeta = mi_ruta.split("/")
    if "datos" in carpeta[-1]:
        print("Estamos en el directorio correcto")
    else:
        mi_ruta = input("Dame la ruta donde esta el fichero: ")
        # nos cambiamos a la ruta del usuario
        os.chdir(mi_ruta)
    print("La ruta de nuestro archivo esta en: ", os.getcwd())
    
    # ruta completa con el archivo
    ubicacion_archivo = mi_ruta + '/' + nombre_fichero

    # abrimos el archivo
    f = open(ubicacion_archivo, encoding = "utf-8")

    # leemos el archivo
    archivo = f.read()

    # mostramos el contenido del archivo
    print(archivo)

    # cerramos el archivo
    f.close()

    # mostramos la cuarta línea del archivo
    f = open(ubicacion_archivo, encoding = "utf-8")
    lineas_archivo = f.readlines()
    
    print(f"La cuarta línea es: '{lineas_archivo[3]}'")

    # cerramos el archivo
    f.close()
```

Archivos XML

Aquí tenéis un archivo llamado peliculas.xml que contiene una lista de películas.

Descarga este fichero para hacer el ejercicio.

En este ejercicio tendréis que crear una función que reciba el nombre del archivo xml y que devuelva lo siguiente:

- Qué tag y atributos tiene el archivo xml.
- La descripción de cada una de las películas que tenemos en ese archivo.
- Los años en que fueron estrenadas las películas.

💡 Pistas 💡

- La función debe recibir un parámetro, el nombre del archivo xml.
- Tendréis que utlizar un bucle for para recorrer todo el archivo xml y extraer la información que os pedimos.
- Recordad el método .text para extraer el texto de un elemento.

Happy coding! 💪

```python
# importamos librería
import xml.etree.ElementTree as ET
```

```python
def informacion_xml(peliculas):
    # abrimos el fichero
    tree = ET.parse(peliculas)
    # elemento que envuelve todo
    root = tree.getroot()
    # printeamos la raiz, el padre
    print(f"tag del archivo {peliculas} =>", root.tag)
    # printeamos los atributos del fichero
    print(f"atributos del archivo {peliculas} => {root.attrib} \n")

    for child in root:
        #genre
        print(f"{child.tag}: es hijo de nuestra raíz: {root.tag}")
        print(f"{child.attrib}: atributos de {child.tag}")
        for subchild in child:
            #decade
            print(f"\t{subchild.tag}: es nieto de {root.tag}")
            print(f"\t{subchild.attrib}: atributos de {subchild.tag}")
            for subsubchild in subchild:
                #movie
                print(f"\t\t{subsubchild.tag}: es bisnieto de {root.tag}")
                print(f"\t\t{subsubchild.attrib}: atributos de {subsubchild.tag}")
                for subsubsubchild in subsubchild:
                    print(f"\t\t\t{subsubsubchild.tag}: es tataranieto de {root.tag}")
                    print(f"\t\t\t{subsubsubchild.attrib}: atributos de {subsubsubchild.tag}")

    # descripcion de las peliculas

    for child in root:
        for subchild in child:
            for subsubchild in subchild:
                print(subsubchild.find('year').text)
                print(subsubchild.find('description').text)
 
```

```python
informacion_xml('peliculas.xml')
```
